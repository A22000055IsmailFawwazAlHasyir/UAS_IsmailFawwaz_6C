```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_supply_kantor = pd.read_csv("/SupplyKantor.csv")
```

```python
df_supply_kantor
```

```python
data_filtered = df_supply_kantor [df_supply_kantor['Jenis']=='Pencil']
data_filtered
```

```python
x = data_filtered['Kota']
y = data_filtered['Jumlah']
plt.figure(figsize=(10, 8))
plt.scatter(x, y, color="green")
plt.xlabel('Kota')
plt.ylabel('Jumlah')

plt.title('1 - Supply Pensil di Setiap Kota')

plt.show()
```

```python
data_filtered2 = df_supply_kantor [df_supply_kantor['Kota']=='Yogyakarta']
data_filtered2
```

```python
x = data_filtered2['Jenis']
y = data_filtered2['Jumlah']
plt.figure(figsize=(10, 8))
plt.scatter(x, y, color="Red")
plt.xlabel('Jenis')
plt.ylabel('Jumlah')

plt.title('2 - Supply Setiap Barang di Kota Yogyakarta')

plt.show()
```
